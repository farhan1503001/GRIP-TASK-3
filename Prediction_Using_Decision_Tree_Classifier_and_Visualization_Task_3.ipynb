

> **Importing Libraries**



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```



> **Importing and preprocessing data**



```python
raw_dataset=load_iris()
feature_set=raw_dataset.data #Creating feature set(x_set)
target_set=raw_dataset.target#Creating label set(y-set)
```



#  **Observing Dataset**


> **Observing Labels of the dataset**





```python
raw_dataset.target_names
```

**So the IRIS dataset is divided into three classes such as Setosa,Versicolor,Virginica so IRIS dataset classification is a three class classification problem**



> **Summary Statistics**



```python
#First converting it into a temporary dataframe
visual_frame=pd.DataFrame(data=raw_dataset.data,columns=raw_dataset.feature_names)
visual_frame.head(10)
```

```python
visual_frame.describe()
```



> **Correlation between Different Features**



```python
visual_frame.corr()
```

# **Splitting Dataset**

```python
x_train,x_test,y_train,y_test=train_test_split(feature_set,target_set,test_size=0.20,shuffle=True)
```



> **Observing the newly created train and test set**



```python
print("Training sample ",x_train[:10])
print("Testing sample ",x_test[:10])
```

# **Applying Decision Tree Classifier**



> **Initializing Decision Tree Classifier**



```python
classifier=DecisionTreeClassifier(criterion='gini',splitter='best',random_state=12)
model=classifier.fit(x_train,y_train)
```



> **Analyzing the performance of the model**



```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_fscore_support
#Training set accuracy
train_accuracy=accuracy_score(y_train,model.predict(x_train))
metric_score=precision_recall_fscore_support(y_train,model.predict(x_train),average=None)#will give us class specific result
print("Training Accuracy is: ",train_accuracy)
print("Category Specific Training Precision,Recall,F-1 Score")
print("Precision\t\t"+"Recall\t\t"+"F-1 Score\t\t")
print(metric_score)
```

**Here, our model's training Accuracy is 100% and training Precision,recall and f-score all are 100%**

**Now for Test Set**

```python
y_pred=model.predict(x_test)
test_accuracy=accuracy_score(y_test,y_pred)
test_metric_score=precision_recall_fscore_support(y_test,y_pred,average=None)#will give us class specific result
print("Testing Accuracy is: ",test_accuracy)
print("Category Specific Testing Precision,Recall,F-1 Score")
print("Precision\t\t"+"Recall\t\t"+"F-1 Score\t\t")
print(test_metric_score)
```

**Here, our model's training Accuracy is 96.66% and training Precision,recall and f-score all are also significant%**

# **Visualization of Decision Tree**


> **Print Text Representation**



```python
from sklearn.tree import export_text
text_tree_represntation=export_text(classifier)
print(text_tree_represntation)
```



> **Plot Tree with plot-tree**



```python
from sklearn.tree import plot_tree
fig=plt.figure(figsize=(16,15))
_=plot_tree(classifier,feature_names=raw_dataset.feature_names,class_names=raw_dataset.target_names,filled=True)
```

```python
fig.savefig('tree_visual.png')
```

```python
from dtreeviz.trees import dtreeviz

visual_representation=dtreeviz(classifier,x_data=feature_set,y_data=target_set,
                               target_name='Target',
                               feature_names=raw_dataset.feature_names,
                               class_names=list(raw_dataset.target_names)
                               )
visual_representation
```

```python
visual_representation.save('graphviz_display.svg')
```

# **Decision Tree Using Entropy**

```python
classifier1=DecisionTreeClassifier(criterion='entropy',splitter='best',random_state=12)
model_entropy=classifier1.fit(x_train,y_train)
print("Model Performance Score/Accuracy: ",model_entropy.score(x_train,y_train))
```

```python
predict_result=classifier.predict(x_test)
print("Testing accuracy for decision tree using entropy: ",accuracy_score(y_test,predict_result))
print("Class Specific Precision, Recall, F-1 Score for Decision tree using entropy")
print(precision_recall_fscore_support(y_test,predict_result))
```
